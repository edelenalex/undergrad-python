Using the techniques demonstrated in Section 10.16, download a data file from <a href="https://www.ncdc.noaa.gov/cag/" target="_blank" rel="noreferrer noopener">NOAA's Climate at a Glance</a> of high (Maximum) temperatures in Louisville, Kentucky for the month of August for all years that the data are available (1948 - 2020).
```python
import pandas as pd
import seaborn as sns

c = lambda f: 5/9 * (f - 32)
temps = [(f, c(f)) for f in range(0, 101, 10)]
temps_df = pd.DataFrame(temps, columns = ['Fahrenheit', 'Celsius'])
axes = temps_df.plot(x = 'Fahrenheit', y = 'Celsius', style = '.-')
y_label = axes.set_ylabel('Celsius')

lou = pd.read_csv('USW00093821-tmax-1-8-1948-2020.csv')

#lou.tail()
lou.columns = ['Date', 'Temperature', 'Anomaly', 'Empty']
lou.Date = lou.Date.floordiv(100)

pd.set_option('precision', 2)
lou.Temperature.describe()

from scipy import stats
linear_regression = stats.linregress(x = lou.Date, y = lou.Temperature)
linear_regression.slope * 2020 + linear_regression.intercept

sns.set_style('whitegrid')
axes = sns.regplot(x = lou.Date, y = lou.Temperature)
axes.set_ylim(10, 70)

year = 2025
slope = linear_regression.slope
intercept = linear_regression.intercept
temperature = slope * year + intercept

print(temperature)
```
